```python
def takeRankValue(dataFrame):
    dictFeature = {}
    for idx, row in dataFrame.iterrows():
        name = row["feature"]
        band = name.split("(")[0]
        if band not in dictFeature:
            dictFeature[band] = 0
        else:
            dictFeature[band] += idx
    return dictFeature
```

```python
dictFeature = takeRankValue(SB)
sortedSB = dict(sorted(dictFeature.items(), key= lambda x:x[1]))
```

```python
dataFrameSB = pd.DataFrame(list(sortedSB.items()), columns= ["feature", "rank"])
```

```python
dataFrameSB.head(20)
```

```python
dictSFFeature = takeRankValue(SF)
```

```python
sortedSF = dict(sorted(dictSFFeature.items(), key= lambda x: x[1]))
```

```python
dataFrameSF = pd.DataFrame(list(sortedSF.items()), columns= ["feature", "rank"])
```

```python
dataFrameSF.head(20)
```

```python
dictREFFeature = takeRankValue(RFE)
```

```python
sortRFE = dict(sorted(dictREFFeature.items(), key= lambda x: x[1] ))
```

```python
dataFrameRFE = pd.DataFrame(sortRFE.items(), columns= ["feature", "rank"])
```

```python
dataFrameRFE
```

```python
fullDataFrame = pd.merge(dataFrameRFE, dataFrameSB, on="feature", how="inner")
fullDataFrame = pd.merge(fullDataFrame, dataFrameSF, on="feature", how= "inner")
```

```python
fullDataFrame["Point"] = fullDataFrame["rank_x"] + fullDataFrame["rank_y"] + fullDataFrame["rank"]
```

```python
fullDataFrame = (fullDataFrame.sort_values(by="Point").reset_index(drop=True))
```
