```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
chunk = pd.read_csv("/sdd/Dubaoset/src/Phong/merged_train.csv", chunksize= 5000000)
```

```python
df = next(chunk)
```

```python
df.dropna()
```

```python
bandType = ['B04B(20)', 'B05B(20)', 'B06B(20)', 'B09B(20)','B10B(20)','B11B(20)','B12B(20)','B14B(20)','B16B(20)','I2B(20)','I4B(20)','IRB(20)','VSB(20)','WVB(20)']
analyst = df.loc[:,bandType]
```

```python
analyst
```

```python
corr = analyst.corr()
```

```python
plt.figure(figsize= (10,9))
sns.heatmap(
    corr,
    annot=True,          # hiển thị giá trị
    fmt=".2f",            # 2 chữ số thập phân
    cmap="RdBu_r",        # màu đối xứng
    vmin=-1, vmax=1,     # chuẩn hóa thang màu
    square=True,         # ô vuông
    linewidths=0.5
)
plt.title("Độ tương quan")
plt.show()
```
