```python
data = pl.read_parquet("/sdd/Dubaoset/src/Phong/Model/data/trainClean/merged_data_part_0.parquet").to_pandas()
```

```python
import numpy as np
```

```python
singleBand = ['B09B','B10B','B11B','B12B','B14B','B16B','I2B','I4B','IRB','WVB']
diffBand= None
fullBand = singleBand + diffBand if diffBand is not None else singleBand
```

```python
bandType = [
    f"{band}_t{i:+d}" if band != "Dem_value" else "Dem_value"
    for i in range(-6, 6)
    for band in fullBand
]
```

```python
from tqdm import tqdm
```

```python
# arr = arr.reshape(-1, 12, len(fullBand))
# -6 -> -1
# -5 -> 0
```

```python
fullResult = []
newCols = [
    f"{band}_t{i:+d}" if band != "Dem_value" else "Dem_value"
    for i in range(-6, 0)
    for band in fullBand
] + ["output_0"]

for i in tqdm(range(0, 6)):
    period = [
        f"{band}_t{i:+d}" if band != "Dem_value" else "Dem_value"
        for i in range(i-6, i)
        for band in fullBand
    ]
    output = [f"output_{i}"]
    # lấy các cột cho 1 sample
    choosenData = data.loc[:, period + output]
    # Chuyển về numpy cho nhẹ 
    fullResult.extend(choosenData.to_numpy())
# sau do phai del data de giam bo nho
# Chuyển ngược về thành dataframe
fullResult = pd.DataFrame(fullResult, columns=newCols)
```
